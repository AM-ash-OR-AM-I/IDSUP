1. An anonymous dataset containing each user’s salary (in dollars) and tenure as a data scientist (in years) is given.
salaries and tenures = [(83000, 8.7), (88000, 8.1), (48000, 0.7), (76000, 6), (69000, 6.5), (76000, 7.5), (60000, 2.5), (83000, 10), (48000, 1.9), (63000, 4.2)]
Find out the average salary for each tenure and print a massage according to its value, i.e. ”less than two”, ”between two and five” and ”more than five” tenure and group together the salaries correspond- ing to each bucket. Compute the average salary for each group.

```python
from collections import defaultdict

salary_tenure = [(83000, 8.7), (88000, 8.1), (48000, 0.7), (76000, 6), (69000, 6.5), (76000, 7.5), (60000, 2.5),
                 (83000, 10), (48000, 1.9), (63000, 4.2)]

salary_bracket = defaultdict(list)
for sal, tenure in salary_tenure:
    if tenure < 2:
        salary_bracket['<2'].append(sal)
    elif 2 <= tenure <= 5:
        salary_bracket['2-5'].append(sal)
    else:
        salary_bracket['5+'].append(sal)

# print(salary_bracket)

for bracket, salaries in salary_bracket.items():
    print(f"{bracket} Y.O.E., avg. sal = {sum(salaries) / len(salaries)}")
```

4.  Write a Python Script to generate random passwords (alphanumeric). Ask users to enter the length ofpassword and number of passwords they want to generate and then print all the generated passwords

```python

import string


def generate_password(length: int):
    universe = ""
    password = ""
    # Ensure at least 1 numeric and alphabet is present
    password += (random.choice(string.ascii_letters) + random.choice(string.digits))
    universe += (string.ascii_letters + string.digits)

    rest = length - len(password)
    for _ in range(rest):
        password += random.choice(universe)

    password = list(password)
    random.shuffle(password)
    password = "".join(password)
    return password


def main():
    length, n = map(int, input("Enter length of and number of passwords (length, n): ").split())
    for i in range(n):
        print(f"Password #{i} = {generate_password(length)}")


if __name__ == "__main__":
    main()
```

6. Read a lists named StringList1 containing strings from the key board. Generate a string MStringList1 that
contains all items of StringList1 that are repeated twice or more number of times and print this list. By
observing the outcome of MStringList1 perform the following tasks:
a. Check wather an item of MStringList1 occurs even number of times or odd number of times in
StringList1.
b. Remove the i th (i ≥ 2) occurrence of a given word in a StringList1.

```python
# Read a lists named StringList1 containing strings from the key board. Generate a string MStringList1 that contains
# all items of StringList1 that are repeated twice or more number of times and print this list. By observing the
# outcome of MStringList1 perform the following tasks: a. Check wather an item of MStringList1 occurs even number of
# times or odd number of times in StringList1. b. Remove the i th (i ≥ 2) occurrence of a given word in a StringList1.


def main():
    string_list1 = input("Enter a list of strings: ").split()
    m_string_list1 = [
        string for string in set(string_list1) if string_list1.count(string) >= 2
    ]
    print(f"MStringList1: {m_string_list1}")

    for string in m_string_list1:
        index = string_list1.index(string)
        # Check if the string occurs more than once, remove i>=2 occurrence
        indexes_to_remove = []
        if string_list1.count(string) > 1:
            remove_index = string_list1.index(string, index + 1)
            indexes_to_remove.append(remove_index)

        for i in indexes_to_remove[::-1]:
            string_list1.pop(i)

    print(f"StringList1 after removing all duplicates of each word: {string_list1}")


main()
```

9. Plot xsinx, x^2 sin(x), x^3 sin(x) and x^4 sin(x) in a single plot in the range x ∈ [−10, 10].

```python
import math
import matplotlib.pyplot as plt

x_values = [i for i in range(-10, 11)]
y1 = [x * math.sin(x) for x in x_values]
y2 = [x ** 2 * math.sin(x) for x in x_values]
y3 = [x ** 3 * math.sin(x) for x in x_values]
y4 = [x ** 4 * math.sin(x) for x in x_values]
plt.plot(x_values, y1)
plt.plot(x_values, y2)
plt.plot(x_values, y3)
plt.plot(x_values, y4)
plt.title('Plots of x sin(x), x^2 sin(x), x^3 sin(x), and x^4 sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Q12. Using the Binomial(n, p) distribution plot a histogram to show the actual binomial samples. Use a line chart to show the normal approximation. Plot both in the same graph. Take n=100, p=0.75 and number of points should be 100.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n = 100
p = 0.75
num_points = 100
binomial_samples = np.random.binomial(n, p, num_points)
binomial_values, binomial_counts = np.unique(binomial_samples, return_counts=True)
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

plt.bar(binomial_values, binomial_counts / num_points, color='skyblue', label='Binomial Samples')

x = np.linspace(0, 100, 1000)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y, color='red', label='Normal Approximation')
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('Binomial Distribution vs Normal Approximation')
plt.legend()
plt.show()
```

# Assignment 2

Plot the Normal PDFs using various value of µ and σ as mentioned below:
µ	σ
0	1
0	2
0	0.5
-1	1

```python
import math

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
mu_values = [0, 0, 0, -1]
sigma_values = [1, 2, 0.5, 1]

for mu, sigma in zip(mu_values, sigma_values):
    y = (1 / (sigma * np.sqrt(2 * math.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    plt.plot(x, y, label=f"µ={mu}, σ={sigma}")

plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Normal PDFs with Various µ and σ')
plt.legend()
plt.show()
```
