e^(x*x) + sin(x) − tan(x) + log(x)

```python

from typing import Callable


def difference_quotient(f: Callable[[float], float], x: float, h: float) -> float:
    """Acts like approximate derivative"""
    return (f(x + h) - f(x)) / h


def f(x):
    return math.exp(x ** 2) + math.sin(x) - math.tan(x) + math.log(x)


difference_quotient(f, 1, 0.000001)
```

3. Write a Python program to find the point of minima of function using Gradient Descent method taking initial solution x0 = 2.
f(x)=x2 +sin(x)

```python
def add(v, w):
    return [vi + wi for vi, wi in zip(v, w)]


def sub(v, w):
    return [vi - wi for vi, wi in zip(v, w)]


def scalar_multiply(c, v):
    return [c * vi for vi in v]


def dot(v, w):
    return sum(vi * wi for vi, wi in zip(v, w))


def sum_of_squares(v):
    return dot(v, v)


def squared_distance(v, w):
    return sum_of_squares(sub(v, w))


def distance(v, w):
    return math.sqrt(squared_distance(v, w))


def gradient_step(v, gradient, step_size: float):
    step = scalar_multiply(step_size, gradient)
    return add(v, step)


def gradient(x):
    return [2 * x[0] + math.cos(x[0])]


v = [-2]
for epoch in range(1000):
    grad = gradient(v)
    v = gradient_step(v, grad, -0.01)
    print(epoch, v)
```

4. Write a Python program to find the point of minima of Rosenbrock function using Gradient Descent method taking initial solution (0,0). Rosenbrock function is defined below.
f(x,y)=(1−x)2 +(y−x2)2

```python
def rosenbrock(v):
    x, y = v
    return (1 - x) ** 2 + (y - x ** 2) ** 2


def rosenbrock_gradient(v):
    x, y = v
    return [-2 * (1 - x) - 4 * x * (y - x ** 2), 2 * (y - x ** 2)]


def gradient_step(v, gradient, step_size):
    return [v_i - step_size * direction_i for v_i, direction_i in zip(v, gradient)]


v = [0, 0]  # initial point
for epoch in range(10000):
    grad = rosenbrock_gradient(v)
    v = gradient_step(v, grad, 0.001)

print("Minima at:", v)
```

### Alternative way

```python
rosenbrock = lambda x, y: (1 - x) ** 2 + (y - x ** 2) ** 2
LR = 0.01


def gradient(f, x, y, h=0.0001):
    return [(f(x + h, y) - f(x, y)) / h, (f(x, h + y) - f(x, y)) / h]


def gradient_step(x, y, dfx, dfy):
    return x - LR * dfx, y - LR * dfy


x = y = 0  # initial point
for epoch in range(10000):
    dfx, dfy = gradient(rosenbrock, x, y)
    x, y = gradient_step(x, y, dfx, dfy)

print("Minima at:", x, y)
```

5. Let X be a binomial random variable with parameters n = 100 and p = 0.6. Find the approximate probability that: 
    1. X lies above 60. 
    2. X lies between 50 and 70.

    using normal approximation to binomial distribution.

```python
def normal_cdf(x, mu, sigma):
    return .5 * (1 + math.erf((x - mu) / (sigma * pow(2, .5))))


def normal_approx_to_binomial(p, n):
    mu = p * n
    sigma = math.sqrt(p * (1 - p) * n)
    return mu, sigma


normal_prob_below = normal_cdf


def normal_prob_above(x, mu, sigma):
    return 1 - normal_prob_below(x, mu, sigma)


def normal_prob_between(mu, sigma, low, high):
    return normal_cdf(high, mu, sigma) - normal_prob_below(low, mu, sigma)


n = 100
p = 0.6
x = 60
low, high = 50, 70
mu, sigma = normal_approx_to_binomial(p, n)
print(normal_prob_above(x, mu, sigma))
print(normal_prob_between(mu, sigma, low, high))
```

```python
import math
from typing import Tuple


def normal_cdf(x, mu, sigma):
    return (1 + math.erf((x - mu) / (sigma * math.sqrt(2)))) / 2


def normal_approximation_to_binomial(n: int, p: float) -> Tuple[float, float]:
    mu = p * n
    sigma = math.sqrt(p * (1 - p) * n)
    return mu, sigma


def normal_probability_above(x, mu, sigma):
    return 1 - normal_cdf(x, mu, sigma)


def normal_probability_between(lo: float, hi: float, mu: float = 0, sigma: float = 1):
    return normal_cdf(hi, mu, sigma) - normal_cdf(lo, mu, sigma)


def normal_probability_outside(lo, hi, mu=0, sigma=1):
    return 1 - normal_probability_between(lo, hi, mu, sigma)


mu, sigma = normal_approximation_to_binomial(100, 0.6)
print(normal_probability_above(60, mu, sigma))
print(normal_probability_between(50, 70, mu, sigma))
```

6. Define p-value and write a python program to find the two-sided p-value with and without continuity correction when the values of x(observed no. of heads), mean and standard deviation are 110, 100, 5 respectively.

```python
normal_probability_below = normal_cdf


def two_sided_p_value(x: float, mu: float, sigma: float = 1) -> float:
    if x >= mu:
        return 2 * normal_probability_above(x, mu, sigma)
    else:
        return 2 * normal_probability_below(x, mu, sigma)


x = 110
mu = 100
sigma = 5
print(two_sided_p_value(x, mu, sigma))
print(two_sided_p_value(x - 0.5, mu, sigma))
```
